## Pergunta de Negócio 1:

### Qual Cidade com Maior Valor de Venda de Produtos da Categoria 'Office Supplies'?

```python
#Filtrando para os ocorrencias de Office Supplies em Categoria
df1 = df_dsa[df_dsa['Categoria'] == 'Office Supplies']

#Agrupando e ordenando (crescente) pelas cidades com o valor_venda
df1 = df1.groupby('Cidade')['Valor_Venda'].sum().sort_values()

#retornando o ultimo indice da ordenacao
df1.idxmax()
```

R - A cidade que obteve maior valor de vendas da categoria Office Supplies foi 'New York City'

## Pergunta de Negócio 2:

### Qual o Total de Vendas Por Data do Pedido?

Demonstre o resultado através de um gráfico de barras.

```python
#agrupando por data e somando os valores de venda
df2 = df_dsa.groupby('Data_Pedido')['Valor_Venda'].sum()

#alterando o tamanho do grafico
plt.figure(figsize = (20,6))
#configurando titulo e eixos do grafico
df2.plot(x = 'Data_Pedido', y = 'Valor_Venda')
plt.title('Total de vendas por data do pedido')
plt.show()
```

## Pergunta de Negócio 3:

### Qual o Total de Vendas por Estado?

Demonstre o resultado através de um gráfico de barras.

```python
#Agrupando por estado e somando os valores de venda
#reset_index usado para reorganizar os indices
df3 = df_dsa.groupby('Estado')['Valor_Venda'].sum().reset_index()

plt.figure(figsize = (20,6))
df3.plot.bar(x = 'Estado', y = 'Valor_Venda', width=0.6, color='red')
plt.title('Total de vendas por estado')
#rotaciona os nomes no eixo x
plt.xticks(rotation=80)

#tight_layout ajusta automaticamente para evitar sobreposicao
plt.tight_layout()
plt.show()
```

## Pergunta de Negócio 4:

### Quais São as 10 Cidades com Maior Total de Vendas?

Demonstre o resultado através de um gráfico de barras.

```python
df4 = df_dsa.groupby('Cidade')['Valor_Venda'].sum().sort_values(ascending=False).head(10).plot.bar(x = 'Cidade', y =  'Valor_Venda', color='green')
plt.title('Top 10 cidades com maior total de vendas')
```

## Pergunta de Negócio 5:

### Qual Segmento Teve o Maior Total de Vendas?

Demonstre o resultado através de um gráfico de pizza.

```python
df5 = df_dsa.groupby('Segmento')['Valor_Venda'].sum().reset_index().sort_values(by = 'Valor_Venda', ascending=False)
```

```python
def valor_total(values):
    def my_format(pct):
        total = sum(values)
        val = int(round(pct * total / 100.0))
        return f'$ {val}'
    return my_format
```

```python
plt.figure(figsize=(16,6))

plt.pie(df5['Valor_Venda'], labels = df5['Segmento'], autopct= valor_total(df5['Valor_Venda']), startangle= 90)

plt.title('Total de vendas por segmento')

```

## Pergunta de Negócio 6 (Desafio Nível Baby):

### Qual o Total de Vendas Por Segmento e Por Ano?

```python
df6 = df_dsa.copy()
#convertendo a data para um tipo adequado
df6['Data_Pedido'] = pd.to_datetime(df6['Data_Pedido'], dayfirst= True)
#criando a coluna ano e colocando o ano correspondente
df6['Ano'] = df6['Data_Pedido'].dt.year

df6_grouped = df6.groupby(['Segmento', 'Ano'])['Valor_Venda'].sum()
df6_grouped
```

## Pergunta de Negócio 7 (Desafio Nível Júnior):

Os gestores da empresa estão considerando conceder diferentes faixas de descontos e gostariam de fazer uma simulação com base na regra abaixo:

- Se o Valor_Venda for maior que 1000 recebe 15% de desconto.
- Se o Valor_Venda for menor que 1000 recebe 10% de desconto.

### Quantas Vendas Receberiam 15% de Desconto?

```python
#cria uma copia do df original
df7 = df_dsa.copy()

#cria a coluna desconto e aplica o valor seguindo a regra proposta
df7['Desconto'] = np.where(df7['Valor_Venda'] > 1000, 0.15, 0.10)
df7.head()
```

```python
#mostrando quantas vendas recebem cada desconto
df7['Desconto'].value_counts()
```

### R - O numero de vendas com 15% de desconto é 457

## Pergunta de Negócio 8 (Desafio Nível Master):

### Considere Que a Empresa Decida Conceder o Desconto de 15% do Item Anterior. Qual Seria a Média do Valor de Venda Antes e Depois do Desconto?

```python
#criando copia do df original
df8 = df7.copy()

#criando coluna do valor da venda com desconto aplicado
df8['Valor_Venda_com_Desconto'] = df8['Valor_Venda'] - (df8['Valor_Venda'] * df8['Desconto']) 
df8.head()
```

```python
#Filtrando o df para obter apenas as vendas que tiveram 14% de desconto
df8_com_desconto15 = df8.query('Desconto == 0.15')
df8_com_desconto15
```

```python
df8_com_desconto15[['Valor_Venda', 'Valor_Venda_com_Desconto']].mean()
```
